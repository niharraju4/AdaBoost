```python
# Getting all different types of features

num_features = [feature for feature in df.columns if df[feature].dtype != '0']
print('Num of numerical Features:', len(num_features))
cat_features = [feature for feature in df.columns if df[feature].dtype == '0']
print('Number of Categroical Features:', len(cat_features))
discrete_features=[feature for feature in df.columns if len(df[feature].unique())<=25]
print('Num of discrete features:', discrete_features)
continious_feature=[feature for feature in num_features if feature not in discrete_features]
print('Num of continious Features:', len(continious_feature))
```
